### Task 1: Daily step counts of two friends - element-wise operations

```python
friend1_steps = np.array([7500, 8200, 9000, 6800, 8500, 7000, 9500])
friend2_steps = np.array([6800, 7900, 8600, 7200, 8000, 7600, 9100])

addition = friend1_steps + friend2_steps
subtraction = friend1_steps - friend2_steps
multiplication = friend1_steps * friend2_steps
division = friend1_steps / friend2_steps

print("Friend 1 steps:", friend1_steps)
print("Friend 2 steps:", friend2_steps)
print("\nAddition (total steps per day):", addition)
print("Subtraction (difference per day):", subtraction)
print("Multiplication:", multiplication)
print("Division (ratio of activity):", division)
```

### Task 2: Spotify recommendation matrix - dot() vs matmul()

```python
# 3x3 matrices: user preference scores and song popularity scores
user_preferences = np.array([
    [0.8, 0.2, 0.5],
    [0.4, 0.9, 0.1],
    [0.3, 0.6, 0.7]
])

song_popularity = np.array([
    [0.6, 0.3, 0.9],
    [0.7, 0.5, 0.2],
    [0.4, 0.8, 0.6]
])

dot_result = np.dot(user_preferences, song_popularity)
matmul_result = np.matmul(user_preferences, song_popularity)

print("Result using dot():")
print(dot_result)
print("\nResult using matmul():")
print(matmul_result)

# For 2D arrays, dot() and matmul() give the same result.
# The difference shows up with arrays of higher dimensions (stacks of matrices):
# matmul() treats extra dimensions as "batches" and multiplies matrix-by-matrix
# within each batch, while dot() does a sum-product over the last axis of the
# first array and the second-to-last axis of the second array, which can mix
# batches in ways that aren't a simple batched matrix multiply.
```

### Task 3: Instagram image pixel brightness - transpose & statistics

```python
pixel_brightness = np.array([
    [120, 150, 200, 90],
    [80, 60, 100, 130],
    [200, 220, 180, 160],
    [50, 70, 90, 110]
])

rotated_image = pixel_brightness.T
print("Original pixel matrix:")
print(pixel_brightness)
print("\nTransposed (rotated) image:")
print(rotated_image)

print("\nMean:", np.mean(pixel_brightness))
print("Median:", np.median(pixel_brightness))
print("Standard deviation:", np.std(pixel_brightness))
print("Variance:", np.var(pixel_brightness))
```

### Task 4: Zomato rating correlation grid - inverse, determinant, eigenvalues/vectors

```python
rating_correlation = np.array([
    [1.0, 0.5, 0.3],
    [0.5, 1.0, 0.4],
    [0.3, 0.4, 1.0]
])

determinant = np.linalg.det(rating_correlation)
print("Determinant:", determinant)

# Only invert if the matrix is invertible (determinant != 0)
if determinant != 0:
    inverse = np.linalg.inv(rating_correlation)
    print("\nInverse matrix:")
    print(inverse)
else:
    print("Matrix is not invertible (determinant is 0). Modify a value and try again.")

eigenvalues, eigenvectors = np.linalg.eig(rating_correlation)
print("\nEigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
```

### Task 5: Swiggy orders - reshape, flatten, split, and stack

```python
# Orders placed in two cities over 6 days, shape (2, 6)
swiggy_orders = np.array([
    [120, 135, 150, 110, 160, 145],
    [90, 100, 95, 105, 115, 120]
])
print("Original array, shape (2, 6):")
print(swiggy_orders)

# Reshape to (3, 4)
reshaped = swiggy_orders.reshape(3, 4)
print("\nReshaped to (3, 4):")
print(reshaped)

# Flatten
flattened = reshaped.flatten()
print("\nFlattened array:")
print(flattened)

# Split into two equal parts
part1, part2 = np.split(flattened, 2)
print("\nPart 1:", part1)
print("Part 2:", part2)

# Stack both parts vertically
stacked = np.vstack((part1, part2))
print("\nVertically stacked:")
print(stacked)
```
