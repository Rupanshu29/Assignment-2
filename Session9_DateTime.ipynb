### Setup

```python
import pandas as pd
import numpy as np
```

### Task 1: Convert a list of delivery timestamp strings to datetime objects

```python
delivery_timestamps = ['2024-06-01 14:30', '2024-06-02 09:15', '2024-06-03 20:45']

delivery_datetimes = pd.to_datetime(delivery_timestamps)

print("Original strings:", delivery_timestamps)
print("\nConverted to datetime:")
print(delivery_datetimes)
```

### Task 2: Flipkart order history CSV - extract year, month, and weekday

```python
# Create a sample Flipkart-style order history CSV
order_dates = pd.date_range(start="2024-01-01", periods=20, freq="17D").strftime("%Y-%m-%d %H:%M:%S")

flipkart_orders = pd.DataFrame({
    "order_id": range(1, 21),
    "order_date": order_dates
})
flipkart_orders.to_csv("flipkart_order_history.csv", index=False)
```

```python
orders_df = pd.read_csv("flipkart_order_history.csv")

# Convert order_date to datetime using the given format
orders_df["order_date"] = pd.to_datetime(orders_df["order_date"], format="%Y-%m-%d %H:%M:%S")

# Extract year, month, and weekday as new columns
orders_df["year"] = orders_df["order_date"].dt.year
orders_df["month"] = orders_df["order_date"].dt.month
orders_df["weekday"] = orders_df["order_date"].dt.day_name()

print(orders_df)
```

### Task 3: Set 'order_date' as index and resample weekly to count total orders per week

```python
orders_indexed = orders_df.set_index("order_date")

weekly_order_counts = orders_indexed.resample("W").size()

print("Total orders placed each week:")
print(weekly_order_counts)
```

### Task 4: Instagram posts - convert 'posted_at' from UTC to Asia/Kolkata timezone

```python
# Create a sample Instagram posts DataFrame with UTC timestamps
posted_at_utc = pd.date_range(start="2024-06-01 06:00", periods=10, freq="3h", tz="UTC")

instagram_posts = pd.DataFrame({
    "post_id": range(1, 11),
    "posted_at": posted_at_utc
})

print("Original 'posted_at' in UTC:")
print(instagram_posts["posted_at"].head())
```

```python
instagram_posts["posted_at_ist"] = instagram_posts["posted_at"].dt.tz_convert("Asia/Kolkata")

print("First 5 timestamps converted to Asia/Kolkata timezone:")
print(instagram_posts[["posted_at", "posted_at_ist"]].head(5))
```

### Task 5: Create an 'is_weekend' feature (True for Saturday/Sunday)

```python
# dayofweek: Monday=0 ... Sunday=6, so Saturday=5 and Sunday=6
orders_df["is_weekend"] = orders_df["order_date"].dt.dayofweek.isin([5, 6])

print(orders_df[["order_date", "weekday", "is_weekend"]])
```
