### Task 1: Line plot of daily steps over the last 7 days - saved as steps_lineplot.png

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
daily_steps = [7500, 8200, 9000, 6800, 8500, 10200, 6000]

plt.figure(figsize=(8, 5))
plt.plot(days, daily_steps, marker='o', color='teal')
plt.title("Daily Steps - Last 7 Days")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig("steps_lineplot.png")
plt.show()

print("Chart saved as steps_lineplot.png")
```

### Task 2: Scatter plot - Zomato rating vs average meal price for 10 restaurants

```python
restaurant_names = [f"Restaurant_{i}" for i in range(1, 11)]
ratings = [4.2, 3.8, 4.5, 4.0, 3.5, 4.1, 3.9, 4.6, 3.7, 4.3]
avg_price = [450, 300, 600, 350, 250, 500, 400, 700, 320, 550]

plt.figure(figsize=(8, 5))
plt.scatter(ratings, avg_price, color='crimson')
plt.title("Zomato Rating vs Average Meal Price")
plt.xlabel("Zomato Rating")
plt.ylabel("Average Meal Price (₹)")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

### Task 3: Bar chart - orders placed on Swiggy, Zomato, and Domino's last month

```python
platforms = ["Swiggy", "Zomato", "Domino's"]
order_counts = [14, 9, 6]
colors = ["#FC8019", "#E23744", "#0078AE"]

plt.figure(figsize=(7, 5))
bars = plt.bar(platforms, order_counts, color=colors, label=platforms)
plt.title("Orders Placed Last Month by Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Orders")
plt.legend(bars, platforms)
plt.show()
```

### Task 4: Histogram - durations (minutes) of last 20 Spotify listening sessions

```python
np.random.seed(42)
session_durations = np.random.randint(10, 120, 20)

plt.figure(figsize=(8, 5))
plt.hist(session_durations, bins=5, color='mediumseagreen', edgecolor='black')
plt.title("Spotify Listening Session Durations (Last 20 Sessions)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

### Task 5: Two subplots - Instagram screen time (line) and posts liked (bar) - saved as social_media_usage.png

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
screen_time_minutes = [90, 120, 75, 110, 130, 160, 140]
posts_liked = [25, 40, 15, 30, 45, 60, 35]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Line plot - daily Instagram screen time
axes[0].plot(days, screen_time_minutes, marker='o', color='purple')
axes[0].set_title("Daily Instagram Screen Time")
axes[0].set_xlabel("Day")
axes[0].set_ylabel("Screen Time (minutes)")
axes[0].grid(True, linestyle='--', alpha=0.5)

# Bar chart - number of posts liked each day
axes[1].bar(days, posts_liked, color='hotpink')
axes[1].set_title("Posts Liked Per Day")
axes[1].set_xlabel("Day")
axes[1].set_ylabel("Number of Posts Liked")

fig.suptitle("Social Media Usage - Last 7 Days")
plt.tight_layout()
plt.savefig("social_media_usage.png")
plt.show()

print("Chart saved as social_media_usage.png")
```
