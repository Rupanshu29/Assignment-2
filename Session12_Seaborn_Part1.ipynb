```python
# Install seaborn if needed (Colab has it pre-installed)
!pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(42)
```

### Task 1: Histplot - distribution of delivery times (minutes) for 50 Zomato orders

```python
delivery_times = np.random.normal(loc=32, scale=8, size=50).round(1)
delivery_times = np.clip(delivery_times, 10, None)  # no negative/unrealistic delivery times

plt.figure(figsize=(8, 5))
sns.histplot(delivery_times, bins=10, color='steelblue')
plt.title("Distribution of Zomato Delivery Times (50 Orders)")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Number of Orders")
plt.show()
```

### Task 2: 'tips' dataset - boxplot of total_bill by day of the week

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x="day", y="total_bill", order=["Thur", "Fri", "Sat", "Sun"])
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

### Task 3: Violinplot - run distributions per IPL team (8 teams), styled with 'darkgrid'

```python
sns.set_theme(style="darkgrid")

teams = ["Mumbai Indians", "Chennai Super Kings", "Royal Challengers Bangalore",
          "Kolkata Knight Riders", "Delhi Capitals", "Sunrisers Hyderabad",
          "Rajasthan Royals", "Punjab Kings"]

ipl_scores_rows = []
for team in teams:
    scores = np.random.normal(loc=170, scale=20, size=14).round().astype(int)
    for score in scores:
        ipl_scores_rows.append({"team": team, "runs": score})

ipl_scores_df = pd.DataFrame(ipl_scores_rows)

plt.figure(figsize=(12, 6))
sns.violinplot(data=ipl_scores_df, x="team", y="runs")
plt.title("Run Distribution per IPL Team")
plt.xlabel("Team")
plt.ylabel("Runs")
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

### Task 4: Countplot - number of songs per genre from 40 Spotify tracks

```python
genres = ["Pop", "Hip-Hop", "Rock", "Electronic"]
song_genres = np.random.choice(genres, size=40, p=[0.35, 0.3, 0.2, 0.15])

spotify_tracks = pd.DataFrame({
    "track_id": range(1, 41),
    "genre": song_genres
})

plt.figure(figsize=(8, 5))
sns.countplot(data=spotify_tracks, x="genre", order=genres, palette="viridis")
plt.title("Number of Songs per Genre (40 Spotify Tracks)")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.show()
```

### Task 5: Kdeplot - distribution of daily step counts for a week, 'whitegrid' theme

```python
sns.set_theme(style="whitegrid")

daily_steps = [7500, 8200, 9000, 6800, 8500, 10200, 6000]

plt.figure(figsize=(8, 5))
sns.kdeplot(daily_steps, fill=True, color='seagreen')
plt.title("Distribution of Daily Step Counts (Fitness App - 1 Week)")
plt.xlabel("Steps")
plt.ylabel("Density")
plt.show()
```
