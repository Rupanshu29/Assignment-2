### Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
sns.set_theme(style="whitegrid")
```

### Task 1: IPL cricket matches - univariate analysis on 'total_runs'

```python
# Create a sample IPL matches dataset
teams = ["Mumbai Indians", "Chennai Super Kings", "Royal Challengers Bangalore",
          "Kolkata Knight Riders", "Delhi Capitals", "Sunrisers Hyderabad"]

ipl_matches = pd.DataFrame({
    "match_id": range(1, 41),
    "team": np.random.choice(teams, 40),
    "total_runs": np.random.normal(loc=175, scale=22, size=40).round().astype(int)
})
ipl_matches.to_csv("ipl_matches.csv", index=False)

ipl_df = pd.read_csv("ipl_matches.csv")
print(ipl_df.head())
```

```python
print("Univariate summary statistics for 'total_runs':")
print(f"Mean:   {ipl_df['total_runs'].mean():.2f}")
print(f"Median: {ipl_df['total_runs'].median():.2f}")
print(f"Min:    {ipl_df['total_runs'].min()}")
print(f"Max:    {ipl_df['total_runs'].max()}")
print(f"Std:    {ipl_df['total_runs'].std():.2f}")
```

### Task 2: Flipkart product reviews - count of reviews per rating (1-5 stars)

```python
categories = ["Electronics", "Fashion", "Home & Kitchen", "Books"]

flipkart_reviews = pd.DataFrame({
    "review_id": range(1, 101),
    "rating": np.random.choice([1, 2, 3, 4, 5], size=100, p=[0.05, 0.10, 0.15, 0.35, 0.35]),
    "category": np.random.choice(categories, size=100)
})

plt.figure(figsize=(7, 5))
sns.countplot(data=flipkart_reviews, x="rating", order=[1, 2, 3, 4, 5], palette="crest")
plt.title("Count of Flipkart Reviews per Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Number of Reviews")
plt.show()
```

### Task 3: Zomato listings - scatterplot of average_cost_for_two vs user_rating

```python
n_restaurants = 60
zomato_listings = pd.DataFrame({
    "restaurant_id": range(1, n_restaurants + 1),
    "average_cost_for_two": np.random.randint(200, 2000, n_restaurants),
    "user_rating": np.round(np.clip(np.random.normal(4.0, 0.5, n_restaurants), 1, 5), 1)
})

plt.figure(figsize=(8, 5))
sns.scatterplot(data=zomato_listings, x="average_cost_for_two", y="user_rating", color="crimson")
plt.title("Average Cost for Two vs User Rating")
plt.xlabel("Average Cost for Two (₹)")
plt.ylabel("User Rating")
plt.show()
```

### Task 4: BookMyShow movies - groupby analysis of average box office collection by genre

```python
genres_list = ["Action", "Drama", "Comedy", "Thriller", "Romance"]

bookmyshow_movies = pd.DataFrame({
    "movie_id": range(1, 31),
    "genre": np.random.choice(genres_list, 30),
    "box_office_collection": np.random.uniform(5, 300, 30).round(2)  # in crores
})

genre_avg_collection = (
    bookmyshow_movies.groupby("genre")["box_office_collection"]
    .mean()
    .sort_values(ascending=False)
    .round(2)
)

print("Average box office collection by genre (₹ Crore), sorted:")
print(genre_avg_collection)
```

### Task 5: Spotify songs - pairplot of danceability, energy, valence, and popularity

```python
n_songs = 80
spotify_songs = pd.DataFrame({
    "danceability": np.random.uniform(0.2, 1.0, n_songs),
    "energy": np.random.uniform(0.1, 1.0, n_songs),
    "valence": np.random.uniform(0.0, 1.0, n_songs),
    "popularity": np.random.randint(10, 100, n_songs)
})

# Make energy and popularity somewhat positively correlated, for a realistic pattern to observe
spotify_songs["popularity"] = (spotify_songs["popularity"] + spotify_songs["energy"] * 40).clip(0, 100)

sns.pairplot(spotify_songs)
plt.suptitle("Pairwise Relationships - Spotify Song Features", y=1.02)
plt.show()
```

**Observation:** Songs with higher `energy` tend to also have higher `popularity` in this data — the points in the energy-vs-popularity panel trend upward together, suggesting more energetic tracks skew more popular. `valence` (musical positivity/happiness), on the other hand, shows no clear relationship with `danceability` or `popularity` here — the points are scattered with no visible trend, meaning how "happy-sounding" a song is doesn't predict how popular or danceable it is in this sample.
