### Setup

```python
import numpy as np
```

### Task 1: Zomato ratings - slice ratings from 2nd and 3rd users

```python
# 4 users x 5 food items ratings (out of 5)
ratings = np.array([
    [4, 5, 3, 2, 4],
    [3, 4, 5, 4, 2],
    [5, 3, 4, 4, 5],
    [2, 4, 3, 5, 4]
])

print("Full ratings array:")
print(ratings)

# Slicing to get rows for the 2nd and 3rd users (index 1 and 2)
user_2_3_ratings = ratings[1:3, :]
print("\nRatings given by the 2nd and 3rd users:")
print(user_2_3_ratings)
```

### Task 2: Boolean indexing - days with steps greater than 8000

```python
steps = np.array([7500, 8200, 9000, 6800, 8500, 7000, 9500, 8100, 6000, 8700])

high_step_days = steps[steps > 8000]
print("Daily steps:", steps)
print("Days with steps greater than 8000:", high_step_days)
```

### Task 3: Fancy indexing - IPL scores from matches 2, 5, and 7

```python
ipl_scores = np.array([145, 178, 162, 190, 205, 155, 172, 188])

# Fancy indexing (matches 2, 5, 7 -> index 1, 4, 6)
selected_scores = ipl_scores[[1, 4, 6]]
print("IPL scores for all 8 matches:", ipl_scores)
print("Scores from matches 2, 5, and 7:", selected_scores)
```

### Task 4: Broadcasting - apply 10% discount to Flipkart prices (no loops)

```python
prices = np.array([999, 1499, 2499, 499, 3999])

# Broadcasting a scalar (0.9) across the whole array to apply a 10% discount
discounted_prices = prices * 0.9

print("Original prices:", prices)
print("Discounted prices (10% off):", discounted_prices)
```

### Task 5: Boolean masking - set negative Spotify ratings to zero

```python
song_ratings = np.array([5, -2, 0, 3, -4, 1, -1, 4])

print("Original ratings:", song_ratings)

# Boolean masking: set all negative values to 0, keep others unchanged
song_ratings[song_ratings < 0] = 0

print("Ratings after masking negatives to 0:", song_ratings)
```
