### Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="whitegrid")
```

### Task 1: Pairplot of the 'tips' dataset - relationships between all numeric variables

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.pairplot(tips)
plt.suptitle("Pairplot of Numeric Variables in the 'tips' Dataset", y=1.02)
plt.show()
```

### Task 2: 'flights' dataset - heatmap of passengers by month and year

```python
flights = sns.load_dataset("flights")
print(flights.head())

# Reshape with pivot_table: months as rows, years as columns, passenger counts as values
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
print(flights_pivot.head())
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Number of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

### Task 3: Relplot on the 'fmri' dataset - signal over time by event type

```python
fmri = sns.load_dataset("fmri")
print(fmri.head())
```

```python
sns.relplot(
    data=fmri, x="timepoint", y="signal",
    hue="event", kind="line", errorbar="sd", height=5, aspect=1.5
)
plt.suptitle("fMRI Signal Over Time by Event Type", y=1.02)
plt.show()
```

### Task 4: 'titanic' dataset - catplot showing survival rate by class, with confidence intervals

```python
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

```python
sns.catplot(
    data=titanic, x="class", y="survived",
    kind="bar", errorbar=("ci", 95), height=5, aspect=1.3, palette="pastel"
)
plt.suptitle("Survival Rate by Passenger Class (with 95% Confidence Intervals)", y=1.02)
plt.show()
```

### Task 5: Jointplot on 'penguins' - bill_length_mm vs flipper_length_mm with regression line

```python
penguins = sns.load_dataset("penguins")
print(penguins.head())
```

```python
sns.jointplot(
    data=penguins, x="bill_length_mm", y="flipper_length_mm",
    kind="reg", height=7, color="darkorange"
)
plt.suptitle("Bill Length vs Flipper Length (with Regression Line)", y=1.02)
plt.show()
```
