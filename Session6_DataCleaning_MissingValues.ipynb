### Setup

```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

### Task 1: IPL player stats CSV - identify missing values with isnull() and notnull()

```python
# Create a sample IPL player stats dataset with some missing values
player_stats = pd.DataFrame({
    "player_name": ["Rohit Sharma", "Virat Kohli", "MS Dhoni", "Jasprit Bumrah",
                     "KL Rahul", "Hardik Pandya", "Ravindra Jadeja", "Shubman Gill",
                     "Rishabh Pant", "Suryakumar Yadav"],
    "runs": [450, 600, np.nan, 20, 480, np.nan, 210, 550, np.nan, 490],
    "matches": [14, 14, 12, np.nan, 13, 14, 14, np.nan, 11, 14],
    "strike_rate": [135.5, 145.2, 130.0, np.nan, 138.7, 150.1, np.nan, 142.3, 148.9, np.nan]
})

player_stats.to_csv("ipl_player_stats.csv", index=False)

# Load it back with pandas
ipl_df = pd.read_csv("ipl_player_stats.csv")
print("IPL player stats:")
print(ipl_df)
```

```python
# Count missing values per column using isnull()
print("Missing values per column (isnull):")
print(ipl_df.isnull().sum())

# Count non-missing (present) values per column using notnull()
print("\nNon-missing values per column (notnull):")
print(ipl_df.notnull().sum())
```

### Task 2: Drop rows with any missing data using dropna(axis=0, how='any')

```python
print("Shape before dropping missing rows:", ipl_df.shape)

ipl_dropped = ipl_df.dropna(axis=0, how='any')

print("Shape after dropping missing rows:", ipl_dropped.shape)
print("\nDataFrame after dropna:")
ipl_dropped
```

### Task 3: Fill missing 'runs' values with the column mean using fillna()

```python
runs_mean = ipl_df["runs"].mean()
print("Mean of 'runs' column:", runs_mean)

ipl_df["runs"] = ipl_df["runs"].fillna(runs_mean)

print("\nUpdated 'runs' column after filling missing values with the mean:")
print(ipl_df["runs"])
```

### Task 4: Zomato-style ratings - forward fill then backward fill

```python
zomato_ratings = pd.DataFrame({
    "restaurant_id": range(1, 9),
    "rating": [4.2, np.nan, np.nan, 3.8, 4.5, np.nan, 4.0, np.nan]
})

print("Ratings before filling:")
print(zomato_ratings)
```

```python
# Forward fill first - propagates the last valid value forward
zomato_ratings["rating_ffill"] = zomato_ratings["rating"].ffill()

# Backward fill for any remaining missing values (e.g. if the dataset starts with NaN)
zomato_ratings["rating_ffill_then_bfill"] = zomato_ratings["rating_ffill"].bfill()

print("Ratings after forward fill, then backward fill for any that remain:")
zomato_ratings
```

### Task 5: Missingness mechanism for a chosen column

**Column chosen:** `matches` in the IPL player stats dataset.

This is most likely **MAR (Missing At Random)** rather than MCAR or MNAR. The number of matches a player has recorded tends to be missing for players who were injured, rested, or added to the squad late in the season — reasons that are related to other observable factors in the dataset (like their role or recent form) rather than being purely random (MCAR), and the missingness isn't directly caused by the value of `matches` itself being high or low (which would be MNAR). In other words, whether `matches` is missing likely depends on other variables we could observe (e.g. injury status, team rotation policy), not on the unseen match count itself.
