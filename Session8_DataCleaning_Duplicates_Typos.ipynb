### Task 1: Zomato order history - find duplicate orders using duplicated()

```python
# Mock Zomato order history CSV with some duplicate orders (same restaurant + same date)
zomato_orders = pd.DataFrame({
    "order_id": range(1, 11),
    "restaurant_name": ["Spice Villa", "Pizza Hub", "Spice Villa", "Curry House",
                         "Pizza Hub", "Burger Point", "Curry House", "Spice Villa",
                         "Sushi Bar", "Burger Point"],
    "order_date": ["2026-06-01", "2026-06-02", "2026-06-01", "2026-06-03",
                    "2026-06-02", "2026-06-04", "2026-06-03", "2026-06-05",
                    "2026-06-06", "2026-06-04"]
})

zomato_orders.to_csv("zomato_order_history.csv", index=False)
orders_df = pd.read_csv("zomato_order_history.csv")
print(orders_df)
```

```python
# Find duplicate orders based on restaurant name and order date
duplicate_mask = orders_df.duplicated(subset=["restaurant_name", "order_date"], keep=False)
duplicate_orders = orders_df[duplicate_mask]

print("Duplicate orders based on restaurant name and date:")
print(duplicate_orders)
```

### Task 2: Flipkart product reviews - top 3 most common duplicate reviews using value_counts()

```python
flipkart_reviews = pd.DataFrame({
    "review_id": range(1, 13),
    "review_text": [
        "Great product, worth the money",
        "Not as expected",
        "Great product, worth the money",
        "Excellent quality!",
        "Not as expected",
        "Great product, worth the money",
        "Average product",
        "Excellent quality!",
        "Not as expected",
        "Excellent quality!",
        "Great product, worth the money",
        "Average product"
    ]
})

review_counts = flipkart_reviews["review_text"].value_counts()
print("Review text counts:")
print(review_counts)

top_3_common_reviews = review_counts.head(3)
print("\nTop 3 most common (duplicate) reviews:")
print(top_3_common_reviews)
```

### Task 3: Spotify playlists - remove exact duplicate rows with drop_duplicates()

```python
spotify_playlists = pd.DataFrame({
    "playlist_name": ["Chill Vibes", "Workout Mix", "Chill Vibes", "Road Trip",
                       "Workout Mix", "Study Beats", "Road Trip", "Chill Vibes"],
    "creator_username": ["asha_m", "rohit22", "asha_m", "priya_k",
                          "rohit22", "karan_v", "priya_k", "asha_m"]
})

print("Before removing duplicates:")
print(spotify_playlists)

cleaned_playlists = spotify_playlists.drop_duplicates()

print("\nAfter removing duplicate playlists:")
print(cleaned_playlists)
```

### Task 4: Instagram usernames - standardize typo variants using replace()

```python
instagram_usernames = pd.DataFrame({
    "user_id": range(1, 7),
    "username": ["insta_queen", "insta-queen", "instaqueen", "insta_queen",
                  "photo_lover", "instaqueen"]
})

print("Before fixing typos:")
print(instagram_usernames)

instagram_usernames["username"] = instagram_usernames["username"].replace(
    {"insta_queen": "instaqueen", "insta-queen": "instaqueen"}
)

print("\nAfter standardizing all variants to 'instaqueen':")
print(instagram_usernames)
```

### Task 5: Paytm-style payment status - unify to 1/0, trimming whitespace and fixing capitalization

```python
payment_status = pd.DataFrame({
    "transaction_id": range(1, 9),
    "status": ["Yes", " yes", "Y", "No", "no ", "N", " Yes ", "NO"]
})

print("Before cleaning:")
print(payment_status)
```

```python
# Trim whitespace and lowercase everything first for consistency
payment_status["status_clean"] = payment_status["status"].str.strip().str.lower()

# Map every paid/unpaid variant to 1/0
status_map = {"yes": 1, "y": 1, "no": 0, "n": 0}
payment_status["status_clean"] = payment_status["status_clean"].map(status_map)

print("After unifying payment status to 1 (paid) / 0 (unpaid):")
print(payment_status)
```
