### Setup

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Task 1: 2x2 grid of subplots - line, bar, scatter, and pie charts

```python
fig, axes = plt.subplots(2, 2, figsize=(11, 9))

# Line chart
x = np.arange(1, 8)
y = [7500, 8200, 9000, 6800, 8500, 10200, 6000]
axes[0, 0].plot(x, y, marker='o', color='teal')
axes[0, 0].set_title("Line Chart - Daily Steps")
axes[0, 0].set_xlabel("Day")
axes[0, 0].set_ylabel("Steps")

# Bar chart
categories = ["Swiggy", "Zomato", "Domino's"]
values = [14, 9, 6]
axes[0, 1].bar(categories, values, color=["#FC8019", "#E23744", "#0078AE"])
axes[0, 1].set_title("Bar Chart - Orders by Platform")
axes[0, 1].set_xlabel("Platform")
axes[0, 1].set_ylabel("Orders")

# Scatter chart
ratings = [4.2, 3.8, 4.5, 4.0, 3.5, 4.1, 3.9, 4.6, 3.7, 4.3]
prices = [450, 300, 600, 350, 250, 500, 400, 700, 320, 550]
axes[1, 0].scatter(ratings, prices, color='crimson')
axes[1, 0].set_title("Scatter Chart - Rating vs Price")
axes[1, 0].set_xlabel("Rating")
axes[1, 0].set_ylabel("Price (₹)")

# Pie chart
genres = ["Action", "Comedy", "Drama", "Thriller"]
genre_share = [35, 25, 20, 20]
axes[1, 1].pie(genre_share, labels=genres, autopct='%1.1f%%',
               colors=["#66c2a5", "#fc8d62", "#8da0cb", "#e78ac3"])
axes[1, 1].set_title("Pie Chart - Movie Genre Share")

plt.tight_layout()
plt.show()
```

### Task 2: Bar chart - average delivery times for Zomato, Swiggy, and Domino's, styled differently

```python
platforms = ["Zomato", "Swiggy", "Domino's"]
avg_delivery_time = [32, 28, 24]

fig, ax = plt.subplots(figsize=(7, 5))
bar_colors = ["#E23744", "#FC8019", "#0078AE"]
bar_edgecolors = ["darkred", "darkorange", "navy"]
bar_linewidths = [1, 3, 5]

for i in range(len(platforms)):
    ax.bar(platforms[i], avg_delivery_time[i], color=bar_colors[i],
           edgecolor=bar_edgecolors[i], linewidth=bar_linewidths[i])

ax.set_title("Average Delivery Time by Platform")
ax.set_xlabel("Platform")
ax.set_ylabel("Average Delivery Time (minutes)")
plt.show()
```

### Task 3: Multi-axis chart - Instagram followers (left) vs average daily posts (right)

```python
influencers = ["Influencer_A", "Influencer_B", "Influencer_C", "Influencer_D", "Influencer_E"]
followers = [120000, 450000, 89000, 300000, 210000]
avg_daily_posts = [2.1, 3.5, 1.2, 2.8, 1.9]

fig, ax1 = plt.subplots(figsize=(9, 5))

color1 = 'steelblue'
ax1.set_xlabel("Influencer")
ax1.set_ylabel("Followers", color=color1)
ax1.plot(influencers, followers, color=color1, marker='o', linestyle='-')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'darkorange'
ax2.set_ylabel("Average Daily Posts", color=color2)
ax2.plot(influencers, avg_daily_posts, color=color2, marker='s', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color2)

fig.suptitle("Instagram Followers vs Average Daily Posts")
fig.tight_layout()
plt.show()
```

### Task 4: Scatter plot - tickets sold vs IMDB rating for 5 Bollywood movies, with annotations

```python
movies = ["Jawan", "Animal", "12th Fail", "Gadar 2", "Rocky Aur Rani"]
tickets_sold_lakhs = [85, 92, 40, 78, 55]
imdb_ratings = [7.1, 6.7, 9.1, 7.0, 7.1]

fig, ax = plt.subplots(figsize=(9, 6))
ax.scatter(imdb_ratings, tickets_sold_lakhs, color='mediumvioletred', s=100)

for i, movie in enumerate(movies):
    ax.annotate(movie, (imdb_ratings[i], tickets_sold_lakhs[i]),
                textcoords="offset points", xytext=(0, 10), ha='center')

ax.set_title("Tickets Sold vs IMDB Rating - Bollywood Movies")
ax.set_xlabel("IMDB Rating")
ax.set_ylabel("Tickets Sold (in lakhs)")
plt.show()
```

### Task 5: Custom annotation on Flipkart monthly sales chart - mark the highest point 'Big Billion Days'

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
monthly_sales_cr = [12, 11, 13, 14, 15, 13, 14, 16, 18, 42, 20, 19]

fig, ax = plt.subplots(figsize=(11, 6))
ax.plot(months, monthly_sales_cr, marker='o', color='#2874F0', linewidth=2)

max_index = monthly_sales_cr.index(max(monthly_sales_cr))
ax.annotate(
    "Big Billion Days",
    xy=(months[max_index], monthly_sales_cr[max_index]),
    xytext=(months[max_index], monthly_sales_cr[max_index] + 6),
    ha='center',
    arrowprops=dict(facecolor='orange', shrink=0.05, width=2, headwidth=8)
)

ax.set_title("Flipkart Monthly Sales")
ax.set_xlabel("Month")
ax.set_ylabel("Sales (₹ Crore)")
plt.show()
```
