### Task 1: ydata-profiling (successor to pandas-profiling) report for a 'Spotify Top 100 Songs' dataset

```python
# Sample Spotify Top 100 dataset (stand-in for the Kaggle CSV) with deliberate
# missing values and a couple of data-type quirks, to make the report meaningful
n = 100
spotify_top100 = pd.DataFrame({
    "track_name": [f"Song_{i}" for i in range(1, n + 1)],
    "artist": np.random.choice(["Artist_A", "Artist_B", "Artist_C", "Artist_D", None],
                                n, p=[0.25, 0.25, 0.25, 0.2, 0.05]),
    "streams_millions": np.round(np.random.uniform(20, 800, n), 2),
    "danceability": np.round(np.random.uniform(0.2, 1.0, n), 3),
    "energy": np.round(np.random.uniform(0.1, 1.0, n), 3),
    "duration_ms": np.random.randint(120000, 300000, n),
    "explicit": np.random.choice([True, False], n)
})
missing_mask = np.random.rand(n) < 0.08
spotify_top100.loc[missing_mask, "streams_millions"] = np.nan
spotify_top100.to_csv("spotify_top100.csv", index=False)

spotify_df = pd.read_csv("spotify_top100.csv")
spotify_df.head()
```

```python
from ydata_profiling import ProfileReport

profile = ProfileReport(spotify_df, title="Spotify Top 100 Songs - Profiling Report", minimal=True)
profile.to_file("spotify_top100_profile.html")
print("Report saved as spotify_top100_profile.html - open it in a browser to view it.")
```

### Task 2: Sweetviz comparison report - Zomato Mumbai vs Zomato Delhi

```python
cuisines = ["North Indian", "Chinese", "Italian", "South Indian", "Continental", "Fast Food"]

zomato_mumbai = pd.DataFrame({
    "restaurant_id": range(1, 51),
    "cuisine": np.random.choice(cuisines, 50),
    "cost_for_two": np.random.randint(300, 2500, 50),
    "rating": np.round(np.random.normal(4.1, 0.4, 50), 1)
})
zomato_mumbai.to_csv("zomato_mumbai.csv", index=False)

zomato_delhi = pd.DataFrame({
    "restaurant_id": range(1, 51),
    "cuisine": np.random.choice(cuisines, 50),
    "cost_for_two": np.random.randint(200, 1800, 50),
    "rating": np.round(np.random.normal(3.8, 0.5, 50), 1)
})
zomato_delhi.to_csv("zomato_delhi.csv", index=False)
```

```python
import sweetviz as sv

mumbai_df = pd.read_csv("zomato_mumbai.csv")
delhi_df = pd.read_csv("zomato_delhi.csv")

comparison_report = sv.compare([mumbai_df, "Mumbai"], [delhi_df, "Delhi"])
comparison_report.show_html("zomato_mumbai_vs_delhi.html", open_browser=False)
print("Comparison report saved as zomato_mumbai_vs_delhi.html")
```

```python
# A quick numeric peek at the same comparison, so you know roughly what to expect in the report
print("Mumbai - cost_for_two mean:", mumbai_df["cost_for_two"].mean().round(2),
      "| rating mean:", mumbai_df["rating"].mean().round(2))
print("Delhi   - cost_for_two mean:", delhi_df["cost_for_two"].mean().round(2),
      "| rating mean:", delhi_df["rating"].mean().round(2))

# YOUR OBSERVATION HERE:
# Open zomato_mumbai_vs_delhi.html and describe one key difference you spot between
# the two cities in the visual report (e.g. cost distribution, rating spread, cuisine mix).
```

### Task 4: pandas-profiling report for Flipkart product reviews - columns needing cleaning

```python
n_reviews = 150
flipkart_reviews = pd.DataFrame({
    "review_id": range(1, n_reviews + 1),
    "product_name": [f"Product_{np.random.randint(1, 40)}" for _ in range(n_reviews)],
    "review_text": np.random.choice(
        ["Great product", "Not as expected", "Excellent quality!", "Average product",
         "Would not recommend", None], n_reviews, p=[0.25, 0.2, 0.2, 0.15, 0.1, 0.1]
    ),
    "rating": np.random.choice([1, 2, 3, 4, 5], n_reviews),
    "reviewer_id": [f"user_{i}" for i in range(1, n_reviews + 1)],  # unique for every row
    "price": [f"₹{np.random.randint(199, 9999)}" for _ in range(n_reviews)]  # stored as text
})

flipkart_profile = ProfileReport(flipkart_reviews, title="Flipkart Reviews - Profiling Report", minimal=True)
flipkart_profile.to_file("flipkart_reviews_profile.html")
print("Report saved as flipkart_reviews_profile.html")
```

```python
# Programmatically check for the same signals the profiling report flags:
# missing values and high-cardinality columns
print("Missing values per column:")
print(flipkart_reviews.isnull().sum())

print("\nUnique value counts per column (high cardinality = close to number of rows):")
print(flipkart_reviews.nunique())
print("Total rows:", len(flipkart_reviews))
```

**Two columns that need cleaning or transformation, based on the report:**

1. **`review_text`** – has missing values (~10% nulls) that need to be handled (dropped or imputed with something like `"No review provided"`) before any text analysis.
2. **`price`** – is stored as an object/string with a `₹` symbol (e.g. `'₹4599'`) instead of a numeric type, so it needs the currency symbol stripped and converting to numeric (`float`/`int`) before it can be used in any calculation or plot.

(`reviewer_id` is also high-cardinality/near-unique, which is expected for an ID column and isn't itself a cleaning issue — profiling tools often flag it, but it doesn't need transforming.)

### Task 5: Auto-EDA on a Swiggy food order CSV - generate a summary report

```python
restaurants = ["Spice Villa", "Pizza Hub", "Curry House", "Burger Point", "Sushi Bar"]

swiggy_orders = pd.DataFrame({
    "order_id": range(1, 121),
    "restaurant": np.random.choice(restaurants, 120),
    "order_amount": np.round(np.random.uniform(120, 1200, 120), 2),
    "delivery_time_min": np.random.randint(15, 60, 120),
    "rating_given": np.random.choice([1, 2, 3, 4, 5, np.nan], 120, p=[0.02,0.03,0.1,0.35,0.4,0.1])
})
swiggy_orders.to_csv("swiggy_orders.csv", index=False)
```

```python
# Using Sweetviz to generate the summary report (chosen tool for this task)
import sweetviz as sv

swiggy_df = pd.read_csv("swiggy_orders.csv")
swiggy_report = sv.analyze(swiggy_df)
swiggy_report.show_html("swiggy_orders_report.html", open_browser=False)

print("Summary report generated: swiggy_orders_report.html")
print("Open this file in a browser as your proof of the generated report.")
```
