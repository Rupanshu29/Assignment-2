
### Task 2: Create a NumPy array of Instagram follower counts

```python
followers = np.array([1200, 15000, 67000, 340000, 1250000])

print("Followers array:", followers)
print("Shape:", followers.shape)
print("Number of dimensions:", followers.ndim)
print("Data type:", followers.dtype)
```

### Task 3: Generate Zomato order IDs using np.arange()

```python
order_ids = np.arange(101, 111)

print("Order IDs:", order_ids)
print("Size:", order_ids.size)
```

### Task 4: Spotify 'like' identity matrix using np.eye()

```python
like_matrix = np.eye(3)

print("Like Identity Matrix:")
print(like_matrix)

# The diagonal values are all 1s, representing a perfect match --
# each song "liking" itself. Off-diagonal 0s mean no relationship
# between two different songs in this matrix.
```

### Task 5: Convert cricket scores list to a NumPy array and check itemsize

```python
cricket_scores = [45, 67, 120, 89, 54]
scores_array = np.array(cricket_scores)

print("Cricket Scores Array:", scores_array)
print("Each score takes", scores_array.itemsize, "bytes in memory")
```
