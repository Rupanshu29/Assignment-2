### Task 4: Large Flipkart product listings Excel file - read in chunks

**Important note:** unlike `pd.read_csv()`, pandas' `pd.read_excel()` does **not** support a `chunksize` parameter for streaming reads — passing `chunksize=` to it raises a `TypeError`. The common workaround is to read the file once, then manually slice it into chunks of the desired size (or use `skiprows`/`nrows` to read chunks directly from disk for very large files). Both approaches are shown below.

```python
# Create a sample Flipkart product listings dataset with 10,000+ rows
categories = ["Electronics", "Fashion", "Home & Kitchen", "Books", "Sports", "Beauty"]
n_products = 10500

flipkart_data = {
    "product_id": range(1, n_products + 1),
    "product_name": [f"Product_{i}" for i in range(1, n_products + 1)],
    "category": [random.choice(categories) for _ in range(n_products)],
    "price": np.round(np.random.uniform(199, 49999, n_products), 2),
    "rating": np.round(np.random.uniform(1, 5, n_products), 1),
    "num_reviews": np.random.randint(0, 5000, n_products)
}
pd.DataFrame(flipkart_data).to_excel("flipkart_products.xlsx", index=False)
```

```python
# Approach 1: read the full file once, then iterate through it in chunks of 2000 rows
flipkart_df = pd.read_excel("flipkart_products.xlsx")

chunk_size = 2000
total_rows = len(flipkart_df)

for start in range(0, total_rows, chunk_size):
    chunk = flipkart_df.iloc[start:start + chunk_size]
    print(f"Chunk starting at row {start}: {len(chunk)} rows")
```

```python
# Approach 2: read directly from disk chunk-by-chunk using nrows/skiprows
# (useful when the file is too large to comfortably hold fully in memory)
chunk_size = 2000
start_row = 0
chunk_num = 1

while start_row < total_rows:
    # skiprows skips the header + already-read rows; header=None since we skip the header row too
    chunk = pd.read_excel(
        "flipkart_products.xlsx",
        skiprows=range(1, start_row + 1),
        nrows=chunk_size
    )
    print(f"Chunk {chunk_num}: {len(chunk)} rows")
    start_row += chunk_size
    chunk_num += 1
```

### Task 5: Semicolon-separated Paytm transactions CSV - detect missing values

```python
# Create a sample Paytm transactions dataset, semicolon-separated, with some missing values
n_txn = 25
paytm_rows = []
for i in range(1, n_txn + 1):
    paytm_rows.append({
        "transaction_id": i,
        "user_id": f"U{1000+i}",
        "amount": round(random.uniform(10, 5000), 2) if random.random() > 0.15 else None,
        "payment_mode": random.choice(["UPI", "Card", "Wallet", "NetBanking", None]),
        "status": random.choice(["Success", "Failed", "Pending"]) if random.random() > 0.1 else None,
        "city": random.choice(cities) if random.random() > 0.2 else None
    })

pd.DataFrame(paytm_rows).to_csv("paytm_transactions.csv", sep=";", index=False)
```

```python
# Read the semicolon-separated CSV
paytm_df = pd.read_csv("paytm_transactions.csv", sep=";")

# Detect missing values
missing_counts = paytm_df.isnull().sum()
columns_with_nulls = missing_counts[missing_counts > 0]

print("Missing value count per column:")
print(missing_counts)

print("\nColumns that have null values:")
print(columns_with_nulls)
```
